```python
from importlib import reload
import honours_project.simulation as simulation
reload(simulation)
import honours_project.data as data
reload(data)

import matplotlib.pyplot as plt
import math
```

```python
cool_seed = 1737125327 # 10, 7

sim = simulation.Simulation(10, 7, shuffle=False, seed=cool_seed)
res = sim.run()

print(sim.seed)
print(res, res.stats["time"])
print(f"Secs / Iteration: {res.stats["time"] / res.iteration}")
```

```python
data = res.stats["util"]

# Calculate the grid size (rows and columns)
num_graphs = len(data)
cols = math.floor(math.sqrt(sim.n))  # Number of columns in the grid
rows = math.ceil(num_graphs / cols)  # Rows based on the number of graphs

# Create the figure
fig, axes = plt.subplots(rows, cols, figsize=(16, 16))  # Adjust figsize as needed
axes = axes.flatten()  # Flatten to easily index subplots

# Plot each utility list in a separate subplot
for i, subdata in enumerate(data):
    axes[i].plot(subdata, label=f'Bidder {i}')
    axes[i].set_title(f'Bidder {i}')
    axes[i].set_xlabel('Response')
    axes[i].set_ylabel('Utility')
    axes[i].legend()

# Hide any empty subplots (if any)
for j in range(num_graphs, len(axes)):
    axes[j].axis('off')

# Adjust layout
plt.tight_layout()

# Display the grid of plots
plt.show()
```
